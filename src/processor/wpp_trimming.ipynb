```python
df_filtered = df[(df['Year'] == 2023) & (df['Type'].isin(["Region", "Subregion", "Country/Area"]))]

df_filtered = df_filtered.rename(columns={"ISO3 Alpha-code": "iso3"})
df_filtered = df_filtered[['iso3',"Region, subregion, country or area *", "Type"]]

print(df_filtered)
```

```python
df_filtered = df_filtered.rename(columns={"Region, subregion, country or area *": "Country"})
df_filtered["Region"] = None
df_filtered["Subregion"] = None

if df_filtered["Type"].str.contains("Region").any():
    df_filtered.loc[df_filtered["Type"] == "Region", "Region"] = df_filtered["Country"]

if df_filtered["Type"].str.contains("Subregion").any():
    df_filtered.loc[df_filtered["Type"] == "Subregion", "Subregion"] = df_filtered["Country"]



df_filtered["Region"] = df_filtered["Region"].fillna(method="ffill")
df_filtered["Subregion"] = df_filtered["Subregion"].fillna(method="ffill")

if df_filtered["Region"].str.contains("Northern America").any():
    df_filtered.loc[df_filtered["Region"] == "Northern America", "Subregion"] = "Northern America"

df_filtered = df_filtered[df["Type"] == "Country/Area"]

df_filtered = df_filtered.drop(columns=["Type"])

print(df_filtered)
print(df_filtered.columns)
```

```python

df_simplified = df_simplified[['iso3', "Year"] + pops]
print(df_simplified)
```

```python
def reshape_column_to_rows(df, fixed_columns, new_col_name="변수", new_val_name="값"):
    return df.melt(id_vars=fixed_columns, var_name=new_col_name, value_name=new_val_name)


df_simplified = reshape_column_to_rows(df_simplified, ['iso3', "Year"] , new_col_name="age", new_val_name="population")
```

```python
# Step 1: Check original values
print(df_simplified["population"].head())

# Step 2: Convert to numeric (coerce errors to NaN)
population_numeric = pd.to_numeric(df_simplified["population"], errors='coerce')
print(population_numeric.head())

# Step 3: Multiply by 1000
population_scaled = population_numeric * 1000
print(population_scaled.head())

# Step 4: Round values
population_rounded = population_scaled.round()
print(population_rounded.head())

# Step 5: Convert to int (nullable if needed)
population_int = population_rounded.astype('Int64')
print(population_int.head())

# Finally, assign back
df_simplified["population"] = population_int
```

```python
%%sql

DROP TABLE IF EXISTS simple;

CREATE TABLE simple (
    iso3 CHAR(3) NOT NULL,
    year CHAR(4) NOT NULL,
    age INT NOT NULL,
    population INT,
    CONSTRAINT pk_simple PRIMARY KEY (iso3, year, age),
    CONSTRAINT fk_region FOREIGN KEY (iso3) REFERENCES region(iso3)
);
```

```python
%%sql

DROP TABLE IF EXISTS region;

CREATE TABLE region (
    iso3 CHAR(3) UNIQUE,
    country VARCHAR(50) NOT NULL,
    region VARCHAR(50) NOT NULL,
    subregion VARCHAR(50) NOT NULL,
    PRIMARY KEY (iso3)
);
```
